```python
def generated_arrays(tiled_node):
    import numpy as np

    for i in range(5):
        exists = f"array_{i}" in tiled_node
        if not exists:
            yield tiled_node.write_array(np.random.random((100, 100)), key=f"array_{i}")
        else:
            yield tiled_node[f"array_{i}"]
```

```python
# Now do tiled client
# from tiled.client import from_uri as tiled_from_uri
# scan_id = c['beamlines/bl531/raw'].keys()[-1]
# raw = c['beamlines/bl531/raw'][scan_id]['primary']['det_image']
raw = c["beamlines/bl531/raw/50febf14-c47f-4de9-94e3-248ab2c050c0/primary/det_image"]
print(raw)
processed_root = c["beamlines/testing"]
random_array_nodes = [array_node for array_node in generated_arrays(processed_root)]

processed1 = random_array_nodes[0]
processed2 = random_array_nodes[1]
extra_processed = random_array_nodes[2]

# create_link detects that items are tiled objects and creates entities if not already existing
# and also creates the link between them
raw_to_processed1 = links.create_link(processed1, "processed_from", raw)
processed1_to_extra = links.create_link(extra_processed, "custom_processed_from", processed1)
extra_processed_to_processed2 = links.create_link(processed2, "processed_from", raw)
```
